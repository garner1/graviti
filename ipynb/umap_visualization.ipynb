```python
import numpy as np
from scipy import ndimage, sparse
from scipy.linalg import eigh, inv, logm, norm
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
import sys
import umap
import warnings
import pandas as pd
import holoviews as hv
import networkx as nx
from holoviews import opts
warnings.filterwarnings('ignore')
```

```python
filename = '/home/garner1/Work/dataset/tissue2graph/pat52_covd_X-XY-RC_data'
mydata = np.load(filename+'.npz',allow_pickle=True)
data = mydata['X'] # log of the descriptor matrix flattened to vector                                                                                                                                                   
nn = int(np.log2(data.shape[0])) #scale the nn logarithmically in the numb of nodes to have enough density of edges for clustering                                                                                    
```

```python
n_neighbors=19; min_dist=0.1; n_components=3; metric='euclidean'
fit = umap.UMAP(
    n_neighbors=n_neighbors,
    min_dist=min_dist,
    n_components=n_components,
    metric=metric
)
u = fit.fit_transform(data[np.random.randint(data.shape[0], size=10000)]);
```

```python
fig = plt.figure(figsize=(10,10))
if n_components == 1:
    ax = fig.add_subplot(111)
    ax.scatter(u[:,0], range(len(u)))
if n_components == 2:
    ax = fig.add_subplot(111)
    ax.scatter(u[:,0], u[:,1])
if n_components == 3:
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(u[:,0], u[:,1], u[:,2], s=10)
```
